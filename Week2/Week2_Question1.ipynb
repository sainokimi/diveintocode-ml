# 問題1：1次関数
$$
y=\frac{1}{2}x+1
$$

```python
import numpy as np
from fractions import Fraction
import matplotlib.pyplot as plt
%matplotlib inline

# -50.0 <= x <= 50.0
x = np.arange(-50.0, 50.1, 0.1)

# Initialize y sequence
y = np.full(len(x), 0.0)
idx = 0

# y = 1/2 * x + 1
for temp_x in x:
    y[idx] = Fraction(1,2)*temp_x + 1
    #print("x = {}, y = {}".format(temp_x, y[idx]))
    idx += 1
```

___
# 問題2：配列の結合

```python
i = range(len(x))
x_con_y = np.insert(y, i, x, 0).reshape(1001, 2)
print(x_con_y)

# Check concatenated x and y correctly
for i in range(5):
    print("x[{}] = {}, y[{}] = {}".format(i, x_con_y[i, 0], i, x_con_y[i, 1]))

for i in range(-6, 0, 1):
    print("x[{}] = {}, y[{}] = {}".format(i, x_con_y[i, 0], i, x_con_y[i, 1]))
```

___
# 問題3：勾配を求める

```python
delta_x = x[1:] - x[:-1]
delta_y = y[1:] - y[:-1]

idx = 0
gradient = []
for temp_x in delta_x:
    gradient.append(delta_y[idx]/temp_x)
    
print(gradient)
```

# 問題5：Pythonの関数化

```python
def compute_gradient(create_seq_func, x_range=(-50, 50.1, 0.1)):
    
    # Create x sequence
    x = np.arange(x_range[0], x_range[1], x_range[2])

    # Create y sequence
    y = create_seq_func(x)

    # Concatenate x and y
    i = range(len(x))
    array_xy = np.insert(y, i, x, 0).reshape(len(x),2)
    
    # Calculate gradient
    x_grad = np.round(x[1:] - x[:-1], 1)
    y_grad = np.round(y[1:] - y[:-1], 1)
    
    idx = 0
    gradient = []
    for temp_x in x_grad:
        gradient.append(y_grad[idx]/temp_x)
        idx += 1

    return array_xy, np.array(gradient)

# y = 1/2 * x + 1
def y_frac1_2_x_plus_1(array_x):

    # Initialize y with same length x
    array_y = np.full(len(array_x), 0.0)
    
    idx = 0
    for temp_x in array_x:
        array_y[idx] = Fraction(1,2)*temp_x + 1
        idx += 1

    return array_y

# y = x^2
def y_x_pow2(array_x):

    # Initialize y with same length x
    array_y = np.full(len(array_x), 0.0)
    
    idx = 0
    for temp_x in array_x:
        array_y[idx] = temp_x**2
        idx += 1

    return array_y

# y = 2 * x^2 * 2^x
def y_2x_pow2_2_powx(array_x):

    # Initialize y with same length x
    array_y = np.full(len(array_x), 0.0)
    
    idx = 0
    for temp_x in array_x:
        array_y[idx] = 2*(temp_x**2)*(2**temp_x)
        idx += 1

    return array_y

# y = sin(x^1/2)
def y_sin_x_pow_frac12(array_x):

    # Initialize y with same length x
    array_y = np.full(len(array_x), 0.0)
    
    idx = 0
    for temp_x in array_x:
        array_y[idx] = np.sin(temp_x**Fraction(1,2))
        idx += 1

    return array_y

array_xy1, gradient1 = compute_gradient(y_frac1_2_x_plus_1)
array_xy2, gradient2 = compute_gradient(y_x_pow2)
array_xy3, gradient3 = compute_gradient(y_2x_pow2_2_powx)
array_xy4, gradient4 = compute_gradient(y_sin_x_pow_frac12, (0.0, 50.1, 0.1))

### Create graph ###

FIGSIZE_X = 20
FIGSIZE_Y = 18

# Create Figure instance for setting figure size
fig = plt.figure(figsize=(FIGSIZE_X, FIGSIZE_Y))

# Arrange graph in two columns
ax_1 = fig.add_subplot(421)
ax_2 = fig.add_subplot(422)
bx_1 = fig.add_subplot(423)
bx_2 = fig.add_subplot(424)
cx_1 = fig.add_subplot(425)
cx_2 = fig.add_subplot(426)
dx_1 = fig.add_subplot(427)
dx_2 = fig.add_subplot(428)

# Set label size
ax_1.tick_params(labelsize=15)
ax_2.tick_params(labelsize=15)
bx_1.tick_params(labelsize=15)
bx_2.tick_params(labelsize=15)
cx_1.tick_params(labelsize=15)
cx_2.tick_params(labelsize=15)
dx_1.tick_params(labelsize=15)
dx_2.tick_params(labelsize=15)

# Set graph title, x-axis label and y-axis label
ax_1.set_title("y = 1/2 * x + 1", fontsize=22)
ax_1.set_xlabel("x", fontsize=15)
ax_1.set_ylabel("y", fontsize=15)
ax_2.set_title("x-gradient", fontsize=22)
ax_2.set_xlabel("x", fontsize=15)
ax_2.set_ylabel("x-gradient", fontsize=15)

bx_1.set_title("y = x^2", fontsize=22)
bx_1.set_xlabel("x", fontsize=15)
bx_1.set_ylabel("y", fontsize=15)
bx_2.set_title("x-gradient", fontsize=22)
bx_2.set_xlabel("x", fontsize=15)
bx_2.set_ylabel("x-gradient", fontsize=15)

cx_1.set_title("y = 2 * x^2 * 2^x", fontsize=22)
cx_1.set_xlabel("x", fontsize=15)
cx_1.set_ylabel("x-gradient", fontsize=15)
cx_2.set_title("x-gradient", fontsize=22)
cx_2.set_xlabel("x", fontsize=15)
cx_2.set_ylabel("x-gradient", fontsize=15)

dx_1.set_title("y = sin(x^1/2)", fontsize=22)
dx_1.set_xlabel("x", fontsize=15)
dx_1.set_ylabel("x-gradient", fontsize=15)
dx_2.set_title("x-gradient", fontsize=22)
dx_2.set_xlabel("x", fontsize=15)
dx_2.set_ylabel("x-gradient", fontsize=15)

# Plot grid
ax_1.grid(which='major', linestyle='--', alpha=0.5)
ax_2.grid(which='major', linestyle='--', alpha=0.5)
bx_1.grid(which='major', linestyle='--', alpha=0.5)
bx_2.grid(which='major', linestyle='--', alpha=0.5)
cx_1.grid(which='major', linestyle='--', alpha=0.5)
cx_2.grid(which='major', linestyle='--', alpha=0.5)
dx_1.grid(which='major', linestyle='--', alpha=0.5)
dx_2.grid(which='major', linestyle='--', alpha=0.5)

# Plot ax - dx
ax_1.plot(array_xy1[:, :1], array_xy1[:, 1:2], lw=3, color="lightcoral")
ax_2.plot(array_xy1[1:, :1], gradient1, lw=3, color="mediumblue")
bx_1.plot(array_xy2[:, :1], array_xy2[:, 1:2], lw=3, color="lightcoral")
bx_2.plot(array_xy2[1:, :1], gradient2, lw=3, color="mediumblue")
cx_1.plot(array_xy3[:, :1], array_xy3[:, 1:2], lw=3, color="lightcoral")
cx_2.plot(array_xy3[1:, :1], gradient3, lw=3, color="mediumblue")
dx_1.plot(array_xy4[:, :1], array_xy4[:, 1:2], lw=3, color="lightcoral")
dx_2.plot(array_xy4[1:, :1], gradient4, lw=3, color="mediumblue")

# Prevent overlap between graphs
plt.tight_layout()

# Display graph
plt.show()

###
```
