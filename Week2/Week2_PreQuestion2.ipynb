# 問題1：乱数の作成

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

mean1 = [-3, 0]
sigma1 = [[1.0, 0.8], [0.8, 1.0]]

# Fix random seed
np.random.seed(0)

values = np.random.multivariate_normal(mean1, sigma1, 500)
print(values[0, 0])
```

___
# 問題2：散布図による可視化

```python
FIGSIZE_X = 10
FIGSIZE_Y = 10

# Create Figure instance for setting figure size
fig = plt.figure(figsize=(FIGSIZE_X, FIGSIZE_Y))

# Arrange graph in two columns
ax = fig.add_subplot(111)

# Set label size
ax.tick_params(labelsize=20)

# Set graph title, x-axis label and y-axis label for ax
ax.set_title("test variance-covariance matrix", fontsize=22)
ax.set_xlabel("x axis", fontsize=18)
ax.set_ylabel("y axis", fontsize=18)

# Plot grid
ax.grid(which='major', linestyle='--', alpha=0.5)

# Plot variance-covariance matrix
plt.scatter(values[:,0], values[:,1], color="lightcoral")

# Display graph
plt.show()

###
```

___
# 問題3：ヒストグラムによる可視化

```python
# Create Figure instance for setting figure size
fig = plt.figure(figsize=(FIGSIZE_X, FIGSIZE_Y))

# Arrange graph in two columns
ax = fig.add_subplot(211)
bx = fig.add_subplot(212)

# Set label size
ax.tick_params(labelsize=20)
bx.tick_params(labelsize=20)

# Set graph title, x-axis label and y-axis label
ax.set_title("test variance-covariance matrix 1", fontsize=22)
ax.set_xlabel("x1 axis", fontsize=18)
ax.set_ylabel("y1 axis", fontsize=18)

bx.set_title("test variance-covariance matrix 2", fontsize=22)
bx.set_xlabel("x2 axis", fontsize=18)
bx.set_ylabel("y2 axis", fontsize=18)

# Line-up xlim
ax.set_xlim(-6, 3, 1)
bx.set_xlim(-6, 3, 1)

# Plot grid
ax.grid(which='major', linestyle='--', alpha=0.5)
bx.grid(which='major', linestyle='--', alpha=0.5)

# Set histgram bins
HIST_BINS = 100

# Plot variance-covariance matrix
ax.hist(values[:,0], bins=HIST_BINS, color="lightcoral")
bx.hist(values[:,1], bins=HIST_BINS, color="mediumblue")
plt.tight_layout()

# Display graph
plt.show()

###
```

# 問題4：データの追加

```python
# Fix random seed
np.random.seed(0)

mean2 = [0, -3]
sigma2 = [[1.0, 0.8], [0.8, 1.0]]

values2 = np.random.multivariate_normal(mean2, sigma2, 500)
print(values2[0, 0])

# Create Figure instance for setting figure size
fig = plt.figure(figsize=(FIGSIZE_X, FIGSIZE_Y))

# Arrange graph in two columns
ax = fig.add_subplot(111)

# Set label size
ax.tick_params(labelsize=20)

# Set graph title, x-axis label and y-axis label for ax
ax.set_title("test variance-covariance matrix", fontsize=22)
ax.set_xlabel("x axis", fontsize=18)
ax.set_ylabel("y axis", fontsize=18)

# Plot grid
ax.grid(which='major', linestyle='--', alpha=0.5)


# Plot variance-covariance matrix
scatter1 = plt.scatter(values[:,0], values[:,1], color="lightcoral")
scatter2 = plt.scatter(values2[:,0], values2[:,1], color="mediumblue")

plt.legend([scatter1, scatter2], ["0", "1"])

# Display graph
plt.show()

###
```

___
# 問題5：データの結合

```python
# Concatenate 'values' and 'values2' vartical
concatenated_values = np.concatenate([values, values2], 0)
print(concatenated_values)
print(concatenated_values.shape)
```

# 問題6：ラベル付け

```python
LABELS = 500

# Create label list
label_zero_ndarray = np.zeros(LABELS, int).reshape(LABELS, 1)
label_ones_ndarray = np.ones(LABELS, int).reshape(LABELS, 1)
labels_ndarray = np.concatenate([label_zero_ndarray, label_ones_ndarray], axis=0)

# Append label column
temp_con_ndarray = np.concatenate([concatenated_values, labels_ndarray], axis=1)
print(temp_con_ndarray)
print(temp_con_ndarray.shape)
```
