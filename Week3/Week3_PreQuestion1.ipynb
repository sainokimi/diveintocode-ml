# 問題6：図の作成

```python
# ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）
sample_count_each_label = df["Species"].value_counts()

print(sample_count_each_label)

### Create graph ###

FIGSIZE_X = 6
FIGSIZE_Y = 6

# Create Figure instance for setting figure size
fig = plt.figure(figsize=(FIGSIZE_X, FIGSIZE_Y))

# Arrange graph in two columns
ax = fig.add_subplot(111)

# Set graph title, x-axis label and y-axis label
ax.set_title("sample count each label", fontsize=22)

# Plot ax
plt.pie(sample_count_each_label, autopct="%1.2f%%", labels=[0, 1, 2], startangle=-30)

# Display graph
plt.show()

###
```

```python
# 特徴量を一つずつ選び、箱ひげ図を使ってラベルごとのデータの分布を可視化する

# Species 0
Species0_attr1 = df.loc[df["Species"] == 0].loc[:, ["sepal length (cm)"]]
Species0_attr2 = df.loc[df["Species"] == 0].loc[:, ["sepal width (cm)"]]
Species0_attr3 = df.loc[df["Species"] == 0].loc[:, ["petal length (cm)"]]
Species0_attr4 = df.loc[df["Species"] == 0].loc[:, ["petal width (cm)"]]

# Species 1
Species1_attr1 = df.loc[df["Species"] == 1].loc[:, ["sepal length (cm)"]]
Species1_attr2 = df.loc[df["Species"] == 1].loc[:, ["sepal width (cm)"]]
Species1_attr3 = df.loc[df["Species"] == 1].loc[:, ["petal length (cm)"]]
Species1_attr4 = df.loc[df["Species"] == 1].loc[:, ["petal width (cm)"]]

# Species 2
Species2_attr1 = df.loc[df["Species"] == 2].loc[:, ["sepal length (cm)"]]
Species2_attr2 = df.loc[df["Species"] == 2].loc[:, ["sepal width (cm)"]]
Species2_attr3 = df.loc[df["Species"] == 2].loc[:, ["petal length (cm)"]]
Species2_attr4 = df.loc[df["Species"] == 2].loc[:, ["petal width (cm)"]]

# Concatenating
attribute1 = pd.concat([Species0_attr1, Species1_attr1, Species2_attr1], axis=1)
attribute2 = pd.concat([Species0_attr2, Species1_attr2, Species2_attr2], axis=1)
attribute3 = pd.concat([Species0_attr3, Species1_attr3, Species2_attr3], axis=1)
attribute4 = pd.concat([Species0_attr4, Species1_attr4, Species2_attr4], axis=1)

attribute1.plot.box(figsize=(7,7))
attribute2.plot.box(figsize=(7,7))
attribute3.plot.box(figsize=(7,7))
attribute4.plot.box(figsize=(7,7))
```

```python
# 特徴量を一つずつ選び、バイオリン図を使ってラベルごとのデータの分布を可視化する

import seaborn as sns

iris_sns = sns.load_dataset("iris")
# print(iris_sns)

fig, (ax, bx, cx, dx) = plt.subplots(1, 4, sharey=True, figsize=(20, 5))

# # sepal_length
sns.violinplot(x="species", y="sepal_length", data=iris_sns, hue="species", ax=ax, inner=None)

# # sepal_width
bx = sns.violinplot(x="species", y="sepal_width", data=iris_sns, hue="species", ax=bx, inner=None)

# # petal_length
cx = sns.violinplot(x="species", y="petal_length", data=iris_sns, hue="species", ax=cx, inner=None)

# # petal_width
dx = sns.violinplot(x="species", y="petal_width", data=iris_sns, hue="species", ax=dx, inner=None)
```

** [箱ひげ図とバイオリン図の利点] **
- 箱ひげ図
 - データの状態が要約され、ひと目で分布のピークがわかる
- バイオリン図
 - データのピークが複数ある場合、それらも確認できる
 - ピークや四分位数以外のデータも、連続的に確認できる

# 問題7：特徴涼感の関係の確認

```python
# 縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る

sns.lmplot(data=iris_sns, x="sepal_length", y="sepal_width", hue="species")
```

```python
# 全ての組み合わせの散布図を一度に表示させる、散布図行列を作る

sns.pairplot(iris_sns, hue="species")
```

```python
# ４つの特徴量に関する相関係数行列を作る

# Calculate correlation coefficient
corr_df = iris_sns.corr(method="pearson")


# 相関係数行列のヒートマップを作る
# Plot heatmap
sns.heatmap(corr_df, vmin=-1.0, vmax=1.0, center=0, annot=True, fmt=".1f")
```
