# 課題：栗まんじゅうが太陽系を覆う日
** 課題の前に調査 **
<br>[太陽系の半径]</br>
<p>太陽系の果ては、冥王星までの距離の約1000倍</p>
<p>冥王星までの距離は約48億km</p>

** メモ **
<p>[球の体積を求める公式]</p>
$$
V=\frac{4}{3}\pi r^3
$$

```python
import math
import matplotlib.pyplot as plt
%matplotlib inline

# Volume a chestnut bun[m3]
CHESTNUT_BUN_VOLUME = 0.07 * 0.07 * 0.05

# Calclulate volume of a sphere
def calc_volume_sphere(radius=1):

    volume = 4/3 * math.pi * radius**3
    
    return volume

# Calculate the time filled up with chestnut bun using vibeine
def calc_time_filled_up(target_rad=1):
    
    list_item_volume = [CHESTNUT_BUN_VOLUME]
    target_volume = calc_volume_sphere(target_rad)
    
    list_time_filled_up = [0]
    
    # Increase item_volume twofold every five minutes
    while target_volume >= list_item_volume[-1]:
        
        list_item_volume.append(list_item_volume[-1]*2)
        list_time_filled_up.append(list_time_filled_up[-1]+5)
        
    return list_item_volume, target_volume, list_time_filled_up

target_radius = float(input("埋め尽くす物体の半径[m]は？ -> "))
list_item_volumes, target_volume, list_time_filled_up = calc_time_filled_up(target_radius)

print("目標物体を栗まんじゅうが埋め尽くす時間は{}分です".format(list_time_filled_up[-1]))

### Create graph ###

FIGSIZE_X = 40
FIGSIZE_Y = 12
X_AXIS_TOP_ADJUST = 5

# Create Figure instance for setting figure size
fig = plt.figure(figsize=(FIGSIZE_X, FIGSIZE_Y))

# Arrange two graphs in two columns
ax = fig.add_subplot(111)

# Set label size
ax.tick_params(labelsize=20)

# Set graph title, x-axis label and y-axis label for ax
ax.set_title("Appearance target filled up with chestnut bun", fontsize=22)
ax.set_xlabel("time[min]", fontsize=18)
ax.set_ylabel("amount chestnut bun", fontsize=18)

# Plot horizontal line as target volume
ax.axhline(target_volume, ls="-.", color="mediumblue")

# Line up y-axis bottom
#ax.set_ylim(YLIM_BOTTOM, list_total_grains[-1]*ADJUST_XLIM_TOP)

# Plot grid
ax.grid(which='major', linestyle='--', alpha=0.5)

# Plot ax and bx
ax.plot(list_time_filled_up, list_item_volumes, linewidth=3, color='lightcoral')

# Display two graphs
plt.show()

###

### ただのメモ（コピペ用） ###
# 太陽系の半径： 4800000000000000 [m]
```
