```python
# 課題1：　べき乗を使用して、計算

t43 = 0.00008 * 2**43
print(t43)
print("厚さ：　{:.2f}キロメートル".format(t43/1000))
```

```python
# 課題2： 単位の変換

print("厚さ：　{:.2f}万キロメートル".format(t43/1000/10000))
```

```python
# 課題3： for文を使用して、課題1を計算

ATUMI = 0.00008
kekka = 0

kekka = ATUMI
for count in range(43):
    kekka *= 2

print(kekka)
print("厚さ：　{:.2f}万キロメートル".format(kekka/1000/10000))
```

```python
# 課題4： 実行速度の測定（べき乗計算）

import time

start = time.time()

# ##
t43 = 0.00008 * 2**43
print(t43)
# ##

elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

```python
# 課題4： 実行速度の測定（for文計算）

import time

start = time.time()

###
ATUMI = 0.00008
kekka = 0

kekka = ATUMI
for count in range(43):
    kekka *= 2

print(kekka)
###

elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

** 課題4の2つの計算方法には、見た目上、大きな違いが出なかった。 **

```python
# 課題5： リストへの保存

import matplotlib.pyplot as plt
%matplotlib inline

ATUMI = 0.00008
kekka = 0
kaisu = 43

# 計算過程を保存するリスト
totyukekka = [0]

kekka = ATUMI
totyukekka[0] = ATUMI
for count in range(kaisu):
    kekka *= 2
    totyukekka.append(kekka)

print(kekka)
print(len(totyukekka))
```

```python
#課題6： グラフの表示

plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(totyukekka)
plt.show()
```

```python
#課題7： グラフのカスタマイズ

import numpy as np
import matplotlib.pyplot as plt2

# グラフサイズを指定したいがために、Figureインスタンスを生成
fig = plt2.figure(figsize=(15, 15))

# 細かい設定をFigureインスタンスを使って設定し、Axesインスタンスを生成（特に、Y軸の設定が欲しかった）
ax = fig.add_subplot(211, title="Increasing thickness", xlabel="number of fold", ylabel="thickness[m]", \
                                              xlim=(-1,44), ylim=(0, (int(totyukekka[-1])*1.05)), \
                                              yticks=(np.arange(0, int(totyukekka[-1]), 100000000)))

# X, Y軸共にラベルの文字サイズを設定
ax.tick_params(labelsize=20)

# グリッド線を表示
ax.grid(which='major', linestyle='--', alpha=0.5)

# 設定したグラフを表示
ax.plot(totyukekka, linewidth=3, color='lightcoral')
```
